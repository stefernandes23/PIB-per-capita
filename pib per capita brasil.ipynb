```python
# localizar em 'fx_idade' o total
total_fx_idade = base.loc[base['fx_idade'] == 'Total']
total_fx_idade
```

regiao + PIB + populacao

Cálculo PIB per capita (é obtido quando dividimos o valor do PIB pelo número de habitantes do município, país ou estado em questão.)

```python
# por não conseguir agrupar já que tinha um valor '...', lambida substitui '...' por 0
total_fx_idade['PIB'] = total_fx_idade['PIB'].apply(lambda x: 0 if x == '...' else x)
```

```python
# astype para int
total_fx_idade['PIB'] = total_fx_idade['PIB'].astype(int)
```

```python
#pegar a media de populacao a partir de 
media_pop = total_fx_idade.groupby(['Granularidade','fx_idade','Regiao'])['Populacao'].mean()
media_pop
```

```python
# dataframe
data = pd.DataFrame(media_pop)
data
```

```python
#agrupamento
dados = data.groupby(['Regiao'])['Populacao'].sum()
dados
```

```python
#dataframe
df = pd.DataFrame(data)
df
```

```python
#agrupamento
df_1 = total_fx_idade.groupby(['Regiao'])['PIB'].sum()
df_1
```

```python
#merge dos df
pibregiao = pd.merge (dados, df_1, on='Regiao')
pibregiao
```

```python
#divisao para resultado per capita
pibregiao['Pib_per_capta'] = pibregiao['PIB']/pibregiao['Populacao']
pibregiao
```

```python
# soma dos dados para ter o total e criar uma linha para isso
total_pop = pibregiao['Populacao'].sum()
total_pib = pibregiao['PIB'].sum()
total_pib_per_capta = total_pib / total_pop  
df_total = pd.DataFrame([[total_pop,total_pib,total_pib_per_capta]] , columns = ['Populacao','PIB','Pib_per_capta'])
df = pd.concat([pibregiao,df_total], ignore_index = False)
df
```

```python
# nomear a coluna 'Regiao' para 'Total'
df.index = ['Centro-Oeste','Nordeste','Norte','Sudeste','Sul','Total']
df
```

```python
#nomear a primeira coluna como 'Regiao'
df.index.name = "Regiao"
df
```

```python
#tirar as 'POPULACAO', 'PIB'
df = df.drop(['Populacao','PIB'], axis = 1)
```

```python
# arrumar multiindex
df = df.set_index('Pib_per_capta', append = True)
df
```
